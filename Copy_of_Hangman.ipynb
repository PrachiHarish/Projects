```python
import random
from more_itertools import locate
import sys
b=input("Do you want to play:")
if b=="yes" and "Yes" and "YES":
    print("welcome to Paandavis Hangman")
    l=["chocolate","icecream","pant","phone", "rose","headphone","muskmelon","instagram", "custardapple","thermometer"]
    g={"chocolate":set(["edible","kids favourite","made from cocoa"]),
      "icecream":set(["cold","summer fav","many flavours"]),
      "pant":set(["clothing","gender fuild clothing","cover legs"]),
      "phone":set(["technology","fast computation","can be hand-held"]),
      "rose":set(["love","red","valentine"]),
      "headphone":set(["technology","ear","sound"]),
      "muskmelon":set(["summer","fruit","watermelon"]),
      "instagram":set(["photes","connect","app"]),
      "custardapple":set(["fruit","sitaphal","green"]),
      "thermometer":set(["temparature","doctor","illness"]),}
    error=['---\n', '---\n'
                  ' |\n'
                  ' |\n','---\n'
                        ' |\n'
                        ' |\n'
                        ' 0\n','---\n'
                                ' |\n'
                                ' |\n'
                                ' 0\n'
                                ' |\n'
                                ' |\n'
                                ' |\n','---\n'
                                      ' |\n'
                                      ' |\n'
                                      ' 0\n'
                                      '/|\\\n'
                                      ' |\n'
                                      ' |\n','---\n'
                                              ' |\n'
                                              ' |\n'
                                              ' 0\n'
                                              '/|\\\n'
                                              ' |\n'
                                              ' |\n'
                                              '/|\\\n',"HANG\n","HANGMAN\n","GAME\n","GAME OVER"]
    a=random.choice(l)
    b=list(a)
    c=list()
    d=10
    s=set()
    collectset=set()
    collectset2=set(b)
    for i in a:
      print("-",end="")
      c.append("-")
    print("\n")
    print("No of chances:",10)
    if len(a)>10:
      for i in range(len(a)):
        if i==2 or i==8:
          c[i]=a[i]
    for i in range(1,11):
      chance=10-i
      if i==3 or i==7 or i==9:
        for j in g[a]-s:
          s.add(j)
          print("Hint:",j)
          break
      for i in c:
        print(i,end="")
      print("\n")
      if collectset==collectset2:
        print("CONGRATULATIONS")
        print(a)
        sys.exit()
      e1=input("Enter:")
      if e1 in b:
        print("correct choice")
        print("No of chances left:",chance)
        collectset.add(e1)
        if collectset==collectset2:
          print("CONGRATULATIONS")
          print(a)
          sys.exit()
        for i in range(len(b)):
          if b[i]==e1:
            c[i]=e1
      elif e1 not in b:
        for k in error:
            print("wrong choice\n")
            print("No of chances left:",chance)
            print(k)
            error.remove(k)
            if k=="GAME OVER":
              sys.exit("Better Luck Next Time")
            break

else:
  print("GOODBYE")

if collectset!=collectset2:
  print("GAME OVER")

```
